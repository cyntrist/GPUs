## Ejercicio 2
* Evaluar el rendimiento **score** para otras redes como VGG19 y ResNet50

```python
import tensorflow as tf
import keras
from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers import Dense, Dropout, GlobalAveragePooling2D

img_rows, img_cols = 32, 32
num_classes = 10

(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Keep a smaller subset so the comparison runs in a reasonable time.
train_limit = 10000
test_limit = 2000
x_train = x_train[:train_limit].astype("float32")
y_train = y_train[:train_limit]
x_test = x_test[:test_limit].astype("float32")
y_test = y_test[:test_limit]

y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)


def build_and_score(backbone_factory, preprocess_fn, name, epochs=2, batch_size=64):
    x_train_p = preprocess_fn(x_train.copy())
    x_test_p = preprocess_fn(x_test.copy())

    try:
        backbone = backbone_factory(
            weights="imagenet",
            include_top=False,
            input_shape=(img_rows, img_cols, 3),
        )
    except Exception:
        backbone = backbone_factory(
            weights=None,
            include_top=False,
            input_shape=(img_rows, img_cols, 3),
        )

    backbone.trainable = False

    model = Sequential([
        backbone,
        GlobalAveragePooling2D(),
        Dropout(0.3),
        Dense(128, activation="relu"),
        Dense(num_classes, activation="softmax"),
    ])

    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    model.fit(
        x_train_p,
        y_train,
        batch_size=batch_size,
        epochs=epochs,
        verbose=1,
        validation_data=(x_test_p, y_test),
    )

    score = model.evaluate(x_test_p, y_test, verbose=0)
    print(f"{name} -> loss: {score[0]:.4f} | accuracy: {score[1]:.4f}")
    return model, score

vgg19_model, vgg19_score = build_and_score(
    keras.applications.VGG19,
    keras.applications.vgg19.preprocess_input,
    "VGG19",
)

resnet50_model, resnet50_score = build_and_score(
    keras.applications.ResNet50,
    keras.applications.resnet50.preprocess_input,
    "ResNet50",
)
```

## Ejercicio 3
* Evaluar el rendimiento **score** cuando se utiliza el dataset CIFAR pero con **100 clases** para las tres redes consideradas: MobileNet, VGG19 y ResNet50
   * El dataset se puede importar con cifar100.load_data()

```python
import tensorflow as tf
import keras
from keras.datasets import cifar100
from keras.models import Sequential
from keras.layers import Dense, Dropout, GlobalAveragePooling2D

img_rows, img_cols = 32, 32
num_classes = 100

(x_train, y_train), (x_test, y_test) = cifar100.load_data(label_mode="fine")

# Keep a smaller subset so the comparison runs in a reasonable time.
train_limit = 10000
test_limit = 2000
x_train = x_train[:train_limit].astype("float32")
y_train = y_train[:train_limit]
x_test = x_test[:test_limit].astype("float32")
y_test = y_test[:test_limit]

y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)


def build_and_score(backbone_factory, preprocess_fn, name, epochs=2, batch_size=64):
    x_train_p = preprocess_fn(x_train.copy())
    x_test_p = preprocess_fn(x_test.copy())

    try:
        backbone = backbone_factory(
            weights="imagenet",
            include_top=False,
            input_shape=(img_rows, img_cols, 3),
        )
    except Exception:
        backbone = backbone_factory(
            weights=None,
            include_top=False,
            input_shape=(img_rows, img_cols, 3),
        )

    backbone.trainable = False

    model = Sequential([
        backbone,
        GlobalAveragePooling2D(),
        Dropout(0.3),
        Dense(128, activation="relu"),
        Dense(num_classes, activation="softmax"),
    ])

    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    model.fit(
        x_train_p,
        y_train,
        batch_size=batch_size,
        epochs=epochs,
        verbose=1,
        validation_data=(x_test_p, y_test),
    )

    score = model.evaluate(x_test_p, y_test, verbose=0)
    print(f"{name} -> loss: {score[0]:.4f} | accuracy: {score[1]:.4f}")
    return model, score

mobilenet_model, mobilenet_score = build_and_score(
    keras.applications.MobileNet,
    keras.applications.mobilenet.preprocess_input,
    "MobileNet",
)

vgg19_model, vgg19_score = build_and_score(
    keras.applications.VGG19,
    keras.applications.vgg19.preprocess_input,
    "VGG19",
)

resnet50_model, resnet50_score = build_and_score(
    keras.applications.ResNet50,
    keras.applications.resnet50.preprocess_input,
    "ResNet50",
)
```
